### Brief Excursion of Gaussians

$$ \mathcal{N}(x|\mu,\,\sigma^{2}) =\frac{1}{\sigma \sqrt{2 \pi}} \mathbf{e}^{-\frac{(x-\mu)^2}{2\sigma^2}} $$ where  
+ $\mu$ is expected value/mean
+ $\sigma$ is standard devation of mean
+ $\sigma^2$ is the variance.

```python
def gaussian(x,mu=0,sigma=0):
    first_term =1/(sigma * np.sqrt(2 * np.pi))
    second_term = np.exp( - (x - mu)**2 / (2 * sigma**2) )
    return  first_term*second_term
```

Let's see how behaves such function on a syntatic data. To this end,
we draw **N** random samples with $\mu=0$ and $\sigma=1$

```python
mu, sigma, N = 0, 1, 100000 # mean and standard deviation
x = np.random.normal(mu, sigma, N)
print('{0:.2f} Mean and {1:.2f} standard deviation of samples'.format(x.mean(),x.std()))
```

```python
_, bins, __ = plt.hist(x, 30, density=True)
plt.plot(bins,gaussian(bins,mu,sigma),linewidth=2, color='r')
plt.title('Gaussian Function')
plt.ylabel('Output')
plt.xlabel('input')
plt.show()
```
