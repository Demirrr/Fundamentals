```python
%matplotlib inline
from d2l import torch as d2l
import torch
```

# Saturation in Sigmoid and Tanh

### Sigmoid
Look at the region where input is greater than 5 or less than -5. We have 0 gradients.


### Tanh
Look at the region where input is greater than 2.5 or less than -2.5. We have 0 gradients.

```python
x = torch.arange(-8.0, 8.0, 0.1, requires_grad=True)
y = torch.sigmoid(x)
y.backward(torch.ones_like(x))

d2l.plot(x.detach().numpy(), [y.detach().numpy(), x.grad.numpy()],
         legend=['sigmoid', 'gradient'], figsize=(4.5, 2.5))
```

```python
x.grad.data.zero_()
y = torch.tanh(x)
y.backward(torch.ones_like(x))

d2l.plot(x.detach().numpy(), [y.detach().numpy(), x.grad.numpy()],
         legend=['tanh', 'gradient'], figsize=(4.5, 2.5))
```

# RELU comes to save us

We have always gradient =1 as long as input greater than 0.0

```python
x.grad.data.zero_()
y = torch.relu(x)
y.backward(torch.ones_like(x))

d2l.plot(x.detach().numpy(), [y.detach().numpy(), x.grad.numpy()],
         legend=['relu', 'gradient'], figsize=(4.5, 2.5))
```
